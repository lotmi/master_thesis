# Notebook for Comparison of Conventionally Fine-Tuned and Tested Models

```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.style.use('ggplot')
```

```python
# Load test data properties

slice_properties = pd.read_csv("/wecare/home/lotte/Thesis/CODE/ETZ/slice_properties.csv", dtype={'scan_nr': int})
testslice_properties = slice_properties[slice_properties['scan_nr'] >= 89]

lesion_properties = pd.read_csv("/wecare/home/lotte/Thesis/CODE/tests_and_exploration/lesion_properties.csv", dtype={'label': int})
lesion_properties["subject_nr"] = [int(subject[3:6]) for subject in lesion_properties['Subject']]
lesion_properties["take"] = [int(subject[-1]) for subject in lesion_properties['Subject']]
lesion_properties["subject_lesion"] = [f'{subject}_l{lesion}' for subject, lesion in zip(lesion_properties['subject_nr'], lesion_properties['label'])] 
testlesion_properties = lesion_properties[lesion_properties['subject_nr'] >= 89]

# Retrieve original quartiles 
lesion_sizes = list(lesion_properties['area'])
q1 = np.quantile(lesion_sizes, .25)
q2 = np.quantile(lesion_sizes, .5)
q3 = np.quantile(lesion_sizes, .75)
print("Lesion size quartiles:", q1, q2, q3) 
```

```python
# Performance per lesion size quartile for YOLOv5
print("Performance per lesion size quartile for YOLOv5")

dr_results = pd.read_csv("/wecare/home/lotte/Thesis/CODE/ETZ/YOLO/YOLOv5/evaluation_runs/base_sampled_s/dr.csv", dtype={'subject': int})
lesion_sizes = [testlesion_properties[testlesion_properties['subject_lesion']==subject_lesion]['area'].item() for subject_lesion in dr_results['subject_lesion']]

size_groups = []
for size in lesion_sizes:
    if size <= q1:
        size_groups.append('q1')
    elif size > q1 and size <= q2:
        size_groups.append('q2')
    elif size > q2 and size <= q3:
        size_groups.append('q3')
    elif size > q3:
        size_groups.append('q4')
dr_results['size'] = lesion_sizes
dr_results['size_group'] = size_groups

q_count = list(dr_results['size_group']).count('q1')
print(q_count, "lesions in Q1")
sub = dr_results[dr_results['size_group']=='q1']
detected = sum(sub['detected_at_least_once'])
dr = detected/q_count
print("DR in Q1:", dr)

q_count = list(dr_results['size_group']).count('q2')
sub = dr_results[dr_results['size_group']=='q2']
print(q_count, "lesions in Q2")
detected = sum(sub['detected_at_least_once'])
dr = detected/q_count
print("DR in Q2:", dr)

q_count = list(dr_results['size_group']).count('q3')
sub = dr_results[dr_results['size_group']=='q3']
print(q_count, "lesions in Q3")
detected = sum(sub['detected_at_least_once'])
dr = detected/q_count
print("DR in Q3:", dr)

q_count = list(dr_results['size_group']).count('q4')
sub = dr_results[dr_results['size_group']=='q4']
print(q_count, "lesions in Q4")
detected = sum(sub['detected_at_least_once'])
dr = detected/q_count
print("DR in Q4:", dr)
```
